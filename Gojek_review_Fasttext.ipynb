```python
import os
import pandas as pd
from tqdm.auto import tqdm

from nltk.tokenize import word_tokenize
from gensim.models import FastText
```

```python
import nltk
nltk.download('punkt')
```

```python
df=pd.read_csv("/content/gojek_review.csv")
df
```

```python
sentences=[word_tokenize(review.lower()) for review in tqdm(df.review)]
sentences[:5]
```

```python
model=FastText(sentences, size=128, window=5, min_count=3, workers=4, iter=100, sg=0, hs=0)
```

```python
os.makedirs("model/fasttext/", exist_ok=True)
```

```python
model.save("model/fasttext/gojek_review.fasttext")
```

```python
model=FastText.load("model/fasttext/gojek_review.fasttext")
```

```python
contoh_data=[
    ['aplikasinya','enak','digunakan','dan','harganya','murah','dan','kompetirif'],
    ['driver','ramah-ramah','komunikatif','murah','senyum','dan','pro-aktif']
]
```

```python
model.train(contoh_data, total_examples=len(contoh_data), epochs=1)
```

```python
model.save("model/fasttext/gojek_review.fastext")
```
