```python
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
from tqdm.auto import tqdm
df=pd.read_csv ("/content/drive/MyDrive/DATASET-TEXT/tokopedia_reviews_dirty.csv")
df.head(5)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
sentences = [word_tokenize(review.lower()) for review in tqdm(df.text)]
sentences [:10]
```

```python
model = Word2Vec (sentences, vector_size=128, window=5, min_count=3, epochs=100, sg=0, hs=0)
```

```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
import umap.umap_ as umap # Make sure umap-learn is installed

# Instantiate the UMAP object
reducer = umap.UMAP()

# Apply fit_transform to your data
X = reducer.fit_transform(w2v.vectors)
df = pd.DataFrame(X, columns=["umap1", "umap2"]) # Corrected column name

# Use index_to_key instead of index2word
df["text"] = w2v.index_to_key # Changed from index2word to index_to_key
```

```python
from umap import UMAP # Import UMAP directly

X = UMAP().fit_transform(w2v.vectors)
```

```python
df = pd.DataFrame(X, columns=["umap1","umap2"])
df["text"] = w2v.index_to_key
```

```python
df
```

```python
fig = px.scatter(df, x="umap1", y="umap2", text="text")
fig.update_traces(textposition='top center')
fig.update_layout(height=2000, title_text='Reduce Word2Vec Visualisation')
fig.show()
```
