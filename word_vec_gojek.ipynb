```python
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
from tqdm.auto import tqdm
df=pd.read_csv ("/content/gojek_review.csv")
df.head()
```

```python
import nltk
nltk.download('punkt')
```

```python
sentences = [word_tokenize(review.lower()) for review in tqdm(df.review)]
sentences [:5]
```

```python
model = Word2Vec (sentences, size=128, window=5, min_count=3, iter=1000, sg=0, hs=0)
```

MENYIMPAN MODEL

```python
os.makedirs ("/content/model/w2v/", exist_ok=True)
```

```python
model.save ("/content/model/w2v/gojek_review.w2v")
```

LOAD MODEL

```python
model=Word2Vec.load("model/w2v/gojek_review.w2v")
```

TRAINING

```python
" layanannya bagus untuk kebutuhan sehari-hari dan kerja".lower() .split()
```

PRE-TRAIN

```python
from umap import UMAP
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
X = UMAP().fit_transform(w2v.vectors)
```

```python
df = pd.DataFrame(X, columns=["umap1","umap2"])
df["text"] = w2v.index2word
```

```python
df
```

```python
fig = px.scatter(df, x="umap1", y="umap2", text="text")
fig.update_traces(textposition='top center')
fig.update_layout(height=800, title_text='Reduce Word2Vec Vicualisation')
fig.show()
```
